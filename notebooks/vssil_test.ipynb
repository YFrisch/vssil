```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
```

```python
sample_h = cv2.imread('/home/yannik/Bilder/master_thesis/samples/vssil_sample_human.jpg')
sample_h = cv2.cvtColor(sample_h, cv2.COLOR_BGR2RGB)
sample_t = cv2.imread('/home/yannik/Bilder/master_thesis/samples/vssil_sample_tiago.jpg')
sample_t = cv2.cvtColor(sample_t, cv2.COLOR_BGR2RGB)
```

```python
font = {'family' : 'Sans',
        'weight' : 'bold',
        'size'   : 22}

matplotlib.rc('font', **font)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(sample_h)
ax[0].set_ylabel('Human')
ax[0].set_yticks([])
ax[0].set_xticks([])
ax[1].imshow(sample_t)
ax[1].set_ylabel('Tiago++')
ax[1].set_yticks([])
ax[1].set_xticks([])
plt.tight_layout()
# plt.show()
fig.savefig('/home/yannik/vssil_human_tiago.svg', bbox_inches='tight')
```
